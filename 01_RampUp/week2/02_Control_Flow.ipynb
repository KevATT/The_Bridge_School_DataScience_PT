## Expresiones Booleanas
Las expresiones booleanas serán evaluadas y devolverán un resultado siempre del tipo `True`o `False`.
Este tipo de dato como ya vimos anteriormente se llama `bool`
En Python existen los siguientes [operadores](https://www.geeksforgeeks.org/relational-operators-in-python/) que representan las expresiones booleanas.

*      `x` == `y`               # x is equal to y
*      `x` != `y`               # x is not equal to y
*      `x` > `y`                # x is greater than y
*      `x` < `y`                # x is less than y
*      `x` >= `y`               # x is greater than or equal to y
*      `x` <= `y`               # x is less than or equal to y






````python
x = 2
y = 4
print(y == (2 * x))
````

````python
# ((not True) and (not False)) > not (A and B) = not A or not B
A = False
B = True
not A and B
````

````python
# not (A or B) = not A and not B
not A or B
````

## IF statement

````python
condition = True
````

````python
if (condition):
  print('la condición es verdadera')
````

````python
perros = 21
gatos = 21

if (perros >= gatos):
  print('Tenemos muchos animales!')
````

````python
condition = False
if (condition): # evaluación por la propia variable si es positiva o negativa
  print('La condición es verdadera')
else:
  print('la condición es falsa')
````

````python
if (perros >= gatos): # estamos evaluando dos variables
  print('Muchos animales')
else:
  print('los gatos son los mejores')
````

````python
condition_1 = True
condition_2 = False

# siempre y cuando unas de las 
# primeras condiciones sea TRUE devolverá un resultado.
if (condition_1):
  print('condición verdadera')
elif (condition_2):
  print('la condición 2 es falsa')
else:
  print('no se cumplen las condiciones')
````

````python
x = 10
y = 5
if x == y:
  print('x and y son iguales')
else:
  if x < y:
    print('x es menor que y')
  else:
    print('x es mayor que y')
````

## `While`

Cuando queremos que la aplicación se ejecute siempre y cuando la condición es positiva, con el `while`ejecutamos al **infinito**

````python
condition = True
# Siempre y cuando `condition` es True, ejecuta el programa...
while condition:
  condition = False # convierte condition en False
````

````python
##También hay que tener cuidado para la recursiva al infinito.
condition = True
# Siempre y cuando `condition` es True, ejecuta el programa...
while condition:
  print('Hola')
````

````python
bus_max_capacidad = 70
pasajeros_actuales = 15

while (pasajeros_actuales < bus_max_capacidad):
  pasajeros_actuales += 1 # esto significa incremento de una unidad
  print('Ha subido un pasajero')
print('El bus está al máximo de la capacidad')
````

## `For`

````python
words = ['gato', 'perro', 'huron', 'pajaro']
for w in words:
  print(w)
````

````python
words[1]
````

````python
counter = 0
while counter < len(words):
  word = words[counter]
  print(word, len(word))
  counter += 1
````

````python
for word in words:
  print(word, len(word))
````
